### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
```

```python
import matplotlib.pyplot as plt

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.bar(lojas, faturamentos, color=['blue', 'green', 'red', 'purple'])
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Vendas por Categoria


```python
categorias_loja1 = loja['Categoria do Produto'].value_counts()
categorias_loja2 = loja2['Categoria do Produto'].value_counts()
categorias_loja3 = loja3['Categoria do Produto'].value_counts()
categorias_loja4 = loja4['Categoria do Produto'].value_counts()
```

```python
# Exemplo para Loja 1 (repita para outras lojas)
top3_loja1 = categorias_loja1.head(3)
plt.pie(top3_loja1, labels=top3_loja1.index, autopct='%1.1f%%')
plt.title('Top 3 Categorias - Loja 1')
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
avaliacao_loja1 = loja['Avaliação da compra'].mean()
avaliacao_loja2 = loja2['Avaliação da compra'].mean()
avaliacao_loja3 = loja3['Avaliação da compra'].mean()
avaliacao_loja4 = loja4['Avaliação da compra'].mean()
```

```python
avaliacoes = [avaliacao_loja1, avaliacao_loja2, avaliacao_loja3, avaliacao_loja4]

plt.bar(lojas, avaliacoes, color=['lightblue', 'lightgreen', 'salmon', 'lavender'])
plt.title('Média de Avaliação por Loja (1-5)')
plt.ylabel('Nota Média')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_loja1 = loja['Produto'].value_counts()
produtos_loja2 = loja2['Produto'].value_counts()
produtos_loja3 = loja3['Produto'].value_counts()
produtos_loja4 = loja4['Produto'].value_counts()
```

```python
top5_loja1 = produtos_loja1.head(5)
plt.barh(top5_loja1.index, top5_loja1.values, color='skyblue')
plt.title('Top 5 Produtos Mais Vendidos - Loja 1')
plt.xlabel('Quantidade Vendida')
plt.show()
```

# 5. Frete Médio por Loja


```python
frete_loja1 = loja['Frete'].mean()
frete_loja2 = loja2['Frete'].mean()
frete_loja3 = loja3['Frete'].mean()
frete_loja4 = loja4['Frete'].mean()
```

```python
fretes = [frete_loja1, frete_loja2, frete_loja3, frete_loja4]

plt.bar(lojas, fretes, color=['orange', 'gold', 'coral', 'lightpink'])
plt.title('Frete Médio por Loja')
plt.ylabel('Custo do Frete (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

# **Conclusão**

```python
# Dados de exemplo (substitua com os valores reais das suas análises)
faturamentos = {
    "Loja 1": 150000,
    "Loja 2": 120000,
    "Loja 3": 180000,
    "Loja 4": 90000
}

avaliacoes = {
    "Loja 1": 4.2,
    "Loja 2": 3.8,
    "Loja 3": 4.5,
    "Loja 4": 3.2
}

fretes_medios = {
    "Loja 1": 35.0,
    "Loja 2": 40.5,
    "Loja 3": 30.2,
    "Loja 4": 50.1
}

# Identificando a melhor e a pior loja
melhor_loja_faturamento = max(faturamentos, key=faturamentos.get)
pior_loja_faturamento = min(faturamentos, key=faturamentos.get)

melhor_loja_avaliacao = max(avaliacoes, key=avaliacoes.get)
pior_loja_avaliacao = min(avaliacoes, key=avaliacoes.get)

melhor_loja_frete = min(fretes_medios, key=fretes_medios.get)
pior_loja_frete = max(fretes_medios, key=fretes_medios.get)

# Contagem de "melhor" e "pior" por loja
from collections import defaultdict
pontuacao = defaultdict(int)

lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]

for loja in lojas:
    if loja == melhor_loja_faturamento:
        pontuacao[loja] += 1
    if loja == pior_loja_faturamento:
        pontuacao[loja] -= 1
    if loja == melhor_loja_avaliacao:
        pontuacao[loja] += 1
    if loja == pior_loja_avaliacao:
        pontuacao[loja] -= 1
    if loja == melhor_loja_frete:
        pontuacao[loja] += 1
    if loja == pior_loja_frete:
        pontuacao[loja] -= 1

# Determinando a loja para vender (menor pontuação)
loja_para_vender = min(pontuacao, key=pontuacao.get)

# Gerando a conclusão automática
print("\n--- CONCLUSÃO BASEADA NOS DADOS ---\n")
print(f"🔥 Melhor desempenho geral: {max(pontuacao, key=pontuacao.get)}")
print(f"❌ Pior desempenho geral: {loja_para_vender}\n")
print("📊 Detalhes:")
print(f"- Melhor faturamento: {melhor_loja_faturamento} (R${faturamentos[melhor_loja_faturamento]:,.2f})")
print(f"- Pior faturamento: {pior_loja_faturamento} (R${faturamentos[pior_loja_faturamento]:,.2f})")
print(f"- Melhor avaliação: {melhor_loja_avaliacao} ({avaliacoes[melhor_loja_avaliacao]}/5)")
print(f"- Pior avaliação: {pior_loja_avaliacao} ({avaliacoes[pior_loja_avaliacao]}/5)")
print(f"- Melhor frete médio: {melhor_loja_frete} (R${fretes_medios[melhor_loja_frete]:.2f})")
print(f"- Pior frete médio: {pior_loja_frete} (R${fretes_medios[pior_loja_frete]:.2f})\n")
print(f"✅ RECOMENDAÇÃO: Vender a {loja_para_vender} e investir nas demais.")
```
